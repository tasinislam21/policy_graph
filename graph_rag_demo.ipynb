````python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import pickle

model = SentenceTransformer('all-MiniLM-L6-v2')

# Choose a text field from your checklist (e.g., 'Description')
texts = df['Description'].fillna("").tolist()
embeddings = model.encode(texts, convert_to_numpy=True)

index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)

faiss.write_index(index, 'gold_dust_faiss.index')
pickle.dump(df.to_dict(), open('gold_dust_metadata.pkl', 'wb'))
````

````python
from neo4j import GraphDatabase

def query_graph(trigger_input):
    uri = "bolt://localhost:7687"
    user = "neo4j"
    password = "Psc1005527" 
    driver = GraphDatabase.driver(uri, auth=(user, password))

    query = """
    WITH toLower($trigger) AS trigger

    MATCH (r:ChecklistRule)
    WHERE toLower(r.TriggerPattern) CONTAINS trigger

    OPTIONAL MATCH (r)-[:CONNECTED_TO]->(s:SOP)
    OPTIONAL MATCH (r)-[:HIGHLIGHTED_BY]->(g:GoldDustChecklist)
    OPTIONAL MATCH (i:Incident)-[:USES_RULE]->(r)
    OPTIONAL MATCH (v:Victim)-[:INVOLVED_IN]->(i)<-[:INVOLVED_IN]-(sus:Suspect)

    RETURN 
      r.TriggerPattern AS rule,
      r.Description AS rule_description,
      s.text AS sop,
      g.text AS gold_dust_item,
      i.description AS incident,
      v.name AS victim,
      sus.name AS suspect
    """

    def format_record(record):
        lines = []
        if record["rule"]: lines.append(f"Trigger: {record['rule']}")
        if record["rule_description"]: lines.append(f"Rule Description: {record['rule_description']}")
        if record["sop"]: lines.append(f"SOP: {record['sop']}")
        if record["gold_dust_item"]: lines.append(f"Gold Dust: {record['gold_dust_item']}")
        if record["incident"]: lines.append(f"Incident: {record['incident']}")
        if record["victim"]: lines.append(f"Victim: {record['victim']}")
        if record["suspect"]: lines.append(f"Suspect: {record['suspect']}")
        return "\n".join(lines)

    with driver.session() as session:
        result = session.run(query, trigger=trigger_input)
        context_records = [format_record(record) for record in result]

    driver.close()
    return context_records
````

````python
from neo4j import GraphDatabase

# Connect to Neo4j instance
uri = "bolt://localhost:7687"  # if using a remote server
username = "neo4j"
password = "Psc1005527"

driver = GraphDatabase.driver(uri, auth=(username, password))

def query_graph(prompt):
    cypher_query = """
    WITH toLower($trigger) AS trigger

    MATCH (r:ChecklistRule)
    WHERE toLower(r.TriggerPattern) CONTAINS trigger

    OPTIONAL MATCH (r)-[:CONNECTED_TO]->(s:SOP)
    OPTIONAL MATCH (r)-[:HIGHLIGHTED_BY]->(g:GoldDustChecklist)
    OPTIONAL MATCH (i:Incident)-[:TRIGGERS]->(r)
    OPTIONAL MATCH (v:Victim)-[:INVOLVED_IN]->(i)<-[:INVOLVED_IN]-(sus:Suspect)

    RETURN 
      r.TriggerPattern AS rule_pattern,
      r.Description AS rule_description,
      s.text AS sop,
      g.text AS gold_dust_item,
      i.description AS incident,
      v.name AS victim,
      sus.name AS suspect
    LIMIT 10
    """

    with driver.session() as session:
        results = session.run(cypher_query, trigger=prompt)
        return [record.data() for record in results]

# Example
prompt = "manipulated"
graph_context = query_graph(prompt)

# Print result
from pprint import pprint
pprint(graph_context)
````

````python
from neo4j import GraphDatabase

uri = "bolt://localhost:7687"
username = "neo4j"
password = "Psc1005527" 

query = """
WITH toLower($trigger) AS trigger 
MATCH (r:ChecklistRule) 
WHERE toLower(r.TriggerPattern) CONTAINS trigger 
OPTIONAL MATCH (r)-[:LINKED_TO_SOP]->(s:SOP)
OPTIONAL MATCH (r)-[:HIGHLIGHTED_BY]->(g:GoldDustChecklist)
OPTIONAL MATCH (i:Incident)-[:TRIGGERS]->(r)
OPTIONAL MATCH (v:Victim)-[:INVOLVED_IN]->(i)<-[:INVOLVED_IN]-(sus:Suspect)
RETURN 
  r.TriggerPattern AS rule_pattern,
  r.Description AS rule_description,
  s.sop_text AS sop,
  g.gold_dust_text AS gold_dust_item,
  i.IncidentNarrative AS incident,
  v.VictimName AS victim,
  sus.SuspectName AS suspect
LIMIT 10
"""

def run_query(trigger_value):
    driver = GraphDatabase.driver(uri, auth=(username, password))
    with driver.session() as session:
        result = session.run(query, trigger=trigger_value)
        for record in result:
            print(record.data())
    driver.close()

run_query("manipulated")
````

````python
from IPython.display import display, Markdown

# Define Cypher queries and expected purposes
queries = [
    {
        "title": "Retrieve full SOP context (for 'stalking')",
        "query": """\
MATCH (s:SOP)<-[:USES_SOP]-(doc:SOPDoc)
OPTIONAL MATCH (r:ChecklistRule)-[:HIGHLIGHTED_BY]->(g:GoldDustChecklist)
OPTIONAL MATCH (r)-[:USES_RULE]->(s)
WHERE toLower(s.name) CONTAINS 'stalking'
RETURN 
  s.name AS SOP,
  collect(DISTINCT r.name) AS RelatedRules,
  collect(DISTINCT g.title) AS HighlightedChecklist,
  collect(DISTINCT doc.content) AS SOPContent;"""
    },
    {
        "title": "Top SOPs by number of associated ChecklistRules",
        "query": """\
MATCH (r:ChecklistRule)-[:USES_RULE]->(s:SOP)
RETURN s.name AS SOP, count(DISTINCT r) AS RuleCount
ORDER BY RuleCount DESC
LIMIT 10;"""
    },
    {
        "title": "Gold Dust Checklist items with most associated rules",
        "query": """\
MATCH (r:ChecklistRule)-[:HIGHLIGHTED_BY]->(g:GoldDustChecklist)
RETURN g.title AS ChecklistItem, count(DISTINCT r) AS RelatedRules
ORDER BY RelatedRules DESC
LIMIT 10;"""
    },
    {
        "title": "Summary of all relationship types and usage count",
        "query": """\
MATCH (r:ChecklistRule)-[rel]->(n)
RETURN type(rel) AS Relationship, count(*) AS Count
ORDER BY Count DESC;"""
    }
]

#shiws the queries in md
for q in queries:
    display(Markdown(f"### {q['title']}\n```cypher\n{q['query']}\n```"))
````
